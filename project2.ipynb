```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder 
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
df=pd.read_csv('data.csv')
df
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns='Y')
y=df['Y']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
ridge_model=Ridge()
lass_model=Lasso()
```

```python
ridge_model.fit(X_train_scaled,y_train)
```

```python
lass_model.fit(X_train_scaled,y_train)
```

```python
y_ridge_predict=ridge_model.predict(X_test_scaled)
```

```python
y_lasso_predict=lass_model.predict(X_test_scaled)
```

```python
mean_squared_error(y_test,y_ridge_predict)
```

```python
mean_squared_error(y_test,y_lasso_predict)
```

```python
r2_score(y_test,y_ridge_predict)
```

```python
import pickle
models={"ridge_model1":ridge_model,"lasso_model1":lass_model}
with open("diabetes_lr_final_model.pkl",'wb') as file :
    pickle.dump((models,scaler),file)
```
